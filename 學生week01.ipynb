**Python 變數名稱**，只能包含字母 **（A-Z, a-z）**、數字 **（0-9）** 和底線 **（_）**，且**不能以數字開頭**。

**不符合規定（非法） 的變數名稱分類與範例：**
1. 不能以數字開頭，變數名稱的第一個字元必須是字母（A-Z, a-z）或底線（_）。
 * ❌ 錯誤範例： **1**st_player, **7**days, **2**026_data
 * ✅ 正確寫法： player_1st, days_7, data_2024

2. 不能包含空格，變數必須是一個連續的字串。
 * ❌ 錯誤範例： user name, total price, my variable
 * ✅ 正確寫法： user_name, total_price (建議使用底線隔開)

3. 不能使用保留字(Keywords)，Python 有一些內建具備特殊功能的關鍵字，不能被當作變數名稱。
 * ❌ 錯誤範例： if, else, while, for, def, class, import, True, False, None, and, or

4. 提示： 如果你在編輯器中看到某個詞變色了（通常是紫色或藍色），那很可能就是保留字。
5. 不能包含特殊符號
除了底線 (_) 之外，其他的標點符號或特殊字元都是不允許的。
 * ❌ 錯誤範例： user-name (被誤認成減法),
 * ❌ price$
 * ❌ email@address
 * ❌ status!
 * ✅ 正確寫法： user_name, price_usd

*   觀念：資料型態 (字串 vs 數字) 的簡單區別。

Python 會自動判斷變數的資料型態，但了解它們的區別很重要。

*   **字串 (String)**：用於儲存文字資料，通常用單引號 `'` 或雙引號 `"` 括起來。
    *   **範例**：`"哈囉世界"`, `'Python 真有趣'`, `name` (來自你的輸入)
    *   **特性**：不能直接進行數學運算。

*   **數字 (Number)**：用於儲存數值資料，主要分為兩種：
    *   **整數 (Integer)**：沒有小數點的數字。
        *   **範例**：`10`, `-5`, `1000`
    *   **浮點數 (Float)**：有小數點的數字。
        *   **範例**：`3.14`, `-0.5`, `99.9`
    *   **特性**：可以進行各種數學運算 (加、減、乘、除等)。
